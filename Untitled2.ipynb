```python
# -*- coding: utf-8 -*-
"""
Spyder Editor

This is a temporary script file.
"""
import numpy as np

class Multilayer_Perceptron:
    
    def __init__(self, w1, b1, w2, b2, w3, b3):
        
        
        self.net={}        
        
        self.net['w1'] = w1
        self.net['b1'] = b1
        
        self.net['w2'] = w2
        self.net['b2'] = b2
        
        self.net['w3'] = w3
        self.net['b3'] = b3        
        
    
    def sigmoid(self, a):
        
        return 1/(1+np.exp(-a))
    
    
    def forward(self, x):
        
        w1, w2, w3 = self.net['w1'] ,self.net['w2'] ,self.net['w3']
        b1, b2, b3 = self.net['b1'] ,self.net['b2'] ,self.net['b3']
        
        a1 = np.dot(x, w1) + b1
        z1 = self.sigmoid(a1)
        
        a2 = np.dot(z1, w2) + b2
        z2 = self.sigmoid(a2)
        
        a3 = np.dot(z2,w3) + b3
        
        return a3
```

## <mark> def forward() :</mark> ##

**For output same size as input**

Input x size = r,c

a1 will be of x dot w1  

`` r x c dot c x ? (=c1) ``

a1 will be of size r x c1

b1 should be of size r x c1

z1 will be same size as a1

<mark> w1 of size c x c1 </mark>

---
a2 will be z1 dot w2  

`` r x c1 dot c1 x ? (=c2) ``

a2 will be of size r x c2

b2 should be of size r x c2

z2 will be same size as a2

<mark> w2 of size c1 x c2 </mark>

---
a3 will be z2 dot w3  

`` r x c2 dot c2 x ? (=c3) ``

a3 will be of size r x c3

b2 should be of size r x c3

<mark> w3 of size c2 x c3 </mark>

**If Final output is same as input then c3 = c
For 3 layers the output size is r x c3 where r is number of rows in x and c3 is #columns in weight vector of last layer (#number of categories)**
